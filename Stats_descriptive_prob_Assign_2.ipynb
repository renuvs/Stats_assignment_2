### Topics: Descriptive Statistics and Probability


```python
#pie char is used for the percent representation
import matplotlib.pyplot as plt

labels = ['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers','Marriott',
          'MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways','Warner-Lambert']

measure_data = [24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00]

#plot the data
plt.figure(figsize =(10, 7))
plt.pie(measure_data, labels=labels, autopct='%.2f')
```

## To find outliers

```python
#we can find outliers by plotting boxplot or histogram. However, boxplot gives a better idea about outliers
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.boxplot(measure_data)
```

As we can see that from the above boxplot the range for valid data is approximately between 10 to 40. Therefore, 'Morgan Stanley' which is 91.36% is an outlier.

## To find average mean, variance and SD

```python
import statistics as stats

print("Mean for given data  = ",round(stats.mean(measure_data),2),
       "\nVariance for given data = ",round(stats.variance(measure_data),2),
       "\nStandard Deviation for given data = ",round(stats.stdev(measure_data),2)
     )
```
